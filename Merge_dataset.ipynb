```python
import os
import json
import re

def basic_cleaning(text):
    text = re.sub(r'(?<!\w)(\d{2,4})-(\d{1,4})(?!\w)', r'\1-\2', text)
    text = re.sub(r'[!\"#$%&\'()*+,\-/:;<=>?@\[\\\]^_`{|}~]', '', text)  # Removed "." from the regex
    return text.strip()

root_path = '/content/adaderana_unzipped/adaderana'  # Make sure this is the correct path
all_files = os.listdir(root_path)
json_files = [f for f in all_files if f.endswith('.json')]
print("Number of JSON files:", len(json_files))

combined_cleaned_lines = []

for jfile in json_files:
    file_path = os.path.join(root_path, jfile)
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)

        # Remove unwanted fields if they exist
        for field_to_drop in ["URL", "Category", "Parent URL"]:
            if field_to_drop in data:
                del data[field_to_drop]

        # Now 'data' only has the fields you want to keep
        text = data.get('News Content', '').strip()
        if text:
            cleaned = basic_cleaning(text)
            if cleaned:
                combined_cleaned_lines.append(cleaned)

print("Done reading. Total lines:", len(combined_cleaned_lines))

# Write out results
with open('adaderana_cleaned.txt', 'w', encoding='utf-8') as outf:
    for line in combined_cleaned_lines:
        outf.write(line + '\n')

print("Cleaned text saved to adaderana_cleaned.txt")
```

```python
import os
import json
import re

def basic_cleaning(text):
    text = re.sub(r'(?<!\w)(\d{2,4})-(\d{1,4})(?!\w)', r'\1-\2', text)
    text = re.sub(r'[!\"#$%&\'()*+,\-/:;<=>?@\[\\\]^_`{|}~]', '', text)  # Removed "." from regex
    return text.strip()

root_path = '/content/Hirunews_unzipped/local-news'
all_files = os.listdir(root_path)
json_files = [f for f in all_files if f.endswith('.json')]
print("Number of JSON files:", len(json_files))

combined_cleaned_lines = []
error_files = []  # Track problematic files

for jfile in json_files:
    file_path = os.path.join(root_path, jfile)

    # Skip empty files
    if os.path.getsize(file_path) == 0:
        error_files.append(jfile)
        continue

    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            data = json.load(f)  # Attempt to parse JSON

        # Remove unwanted fields
        for field_to_drop in ["URL", "Category", "Parent URL"]:
            if field_to_drop in data:
                del data[field_to_drop]

        # Extract and clean text
        text = data.get('News Content', '').strip()
        if text:
            cleaned = basic_cleaning(text)
            if cleaned:
                combined_cleaned_lines.append(cleaned)

    except json.JSONDecodeError:  # Handle corrupt JSON files
        error_files.append(jfile)

print(f"Done reading. Total valid lines: {len(combined_cleaned_lines)}")
print(f"Skipped {len(error_files)} problematic files.")

# Save cleaned text
with open('Hirunews_cleaned.txt', 'w', encoding='utf-8') as outf:
    for line in combined_cleaned_lines:
        outf.write(line + '\n')

print("Cleaned text saved to Hirunews_cleaned.txt")
```

```python
import re

# 1) Define the function to remove Tamil characters only.
def remove_tamil_characters(text):
    """
    Removes any characters in the Tamil Unicode block (U+0B80–U+0BFF).
    Returns the modified string with all Tamil characters stripped out.
    """
    return re.sub(r'[\u0B80-\u0BFF]', '', text)

# 2) Define your source file (the combined corpus) and the output file
INPUT_FILE = 'sinhala_corpus_cleaned_all.txt'
OUTPUT_FILE = 'sinhala_corpus_cleaned_no_tamil.txt'

# 3) Read the input file, process each line, and collect results
filtered_lines = []
with open(INPUT_FILE, 'r', encoding='utf-8') as infile:
    for line in infile:
        line = line.strip()
        if not line:
            continue  # skip empty lines

        # Remove Tamil characters
        cleaned_line = remove_tamil_characters(line)

        # Optionally, if a line becomes empty after removing Tamil, you might skip it:
        # if not cleaned_line.strip():
        #     continue

        filtered_lines.append(cleaned_line)

# 4) Write the filtered lines to a new file
with open(OUTPUT_FILE, 'w', encoding='utf-8') as outfile:
    for line in filtered_lines:
        outfile.write(line + "\n")

print(f"Done. Processed {len(filtered_lines)} lines from {INPUT_FILE}.")
print(f"Output file saved as {OUTPUT_FILE}.")
```

```python
import re

def keep_sinhala_and_common_ascii(text):
    """
    Removes everything except:
      - Sinhala block: \u0D80-\u0DFF
      - ASCII letters: a-zA-Z
      - Digits: 0-9
      - Basic punctuation: . , ! ? : ; ' " - and whitespace
    Feel free to add or remove characters from the allowed set below.
    """
    # This is the regex to keep only allowed characters
    # Modify this as needed to suit your exact requirements
    allowed_chars_pattern = r'[a-zA-Z0-9\u0D80-\u0DFF.,!?;:\'" -]+'

    # Find all matches of the allowed characters and join them back into a string
    cleaned_text = ''.join(re.findall(allowed_chars_pattern, text))

    return cleaned_text  # Return the cleaned text

INPUT_FILE = "sinhala_corpus_cleaned_no_tamil.txt"
OUTPUT_FILE = "sinhala_only.txt"

filtered_lines = []
with open(INPUT_FILE, 'r', encoding='utf-8') as infile:
    for line in infile:
        line = line.strip()
        if not line:
            continue

        # Keep only Sinhala + ASCII per the function above
        cleaned_line = keep_sinhala_and_common_ascii(line)

        # Optionally skip if it becomes empty after cleaning
        # if not cleaned_line.strip():
        #     continue

        filtered_lines.append(cleaned_line)

# Write out the cleaned lines
with open(OUTPUT_FILE, 'w', encoding='utf-8') as outfile:
    for l in filtered_lines:
        outfile.write(l + "\n")

print(f"Processed {len(filtered_lines)} lines from {INPUT_FILE}.")
print(f"Output saved to {OUTPUT_FILE}.")
```

```python
import re
import unicodedata

def normalize_text(text):
    """
    Applies Unicode normalization (NFKC) and removes extra whitespace.
    """
    text = unicodedata.normalize('NFKC', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

def split_sentences(text):
    """
    Splits text into sentences using punctuation as delimiters.
    Adjust the regex if needed (e.g., to include Sinhala-specific punctuation).
    """
    # This regex splits on a period, exclamation, or question mark,
    # followed by one or more whitespace characters.
    sentences = re.split(r'(?<=[.?!])\s+', text)
    # Remove any empty sentences and extra spaces
    return [s.strip() for s in sentences if s.strip()]

# Define your input and output files
INPUT_FILE = '/content/drive/MyDrive/sinhala_only.txt'
OUTPUT_FILE = 'sinhala_corpus_normalized.txt'

all_sentences = []

# Read the normalized file, split each line into sentences, and collect them.
with open(INPUT_FILE, 'r', encoding='utf-8') as infile:
    for line in infile:
        line = line.strip()
        if not line:
            continue  # Skip empty lines

        # Normalize the line (if not already normalized)
        normalized_line = normalize_text(line)
        sentences = split_sentences(normalized_line)
        all_sentences.extend(sentences)

# Write each sentence to a new file, one sentence per line.
with open(OUTPUT_FILE, 'w', encoding='utf-8') as outfile:
    for sentence in all_sentences:
        outfile.write(sentence + "\n")

print(f"Processed {len(all_sentences)} sentences from {INPUT_FILE} and saved them to {OUTPUT_FILE}.")
```

```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
import unicodedata
from collections import Counter
import matplotlib.font_manager as fm

# Set Seaborn style
sns.set_style("whitegrid")

# Define file path
CORPUS_FILE = "/content/drive/MyDrive/sinhala_only.txt"

# Set Sinhala font path (Adjust path if needed)
SINHALA_FONT_PATH = "/usr/share/fonts/truetype/noto/NotoSansSinhala-Regular.ttf"

# Check if the font exists, otherwise use default
if os.path.exists(SINHALA_FONT_PATH):
    sin_font = fm.FontProperties(fname=SINHALA_FONT_PATH)
else:
    print("[WARNING] Sinhala font not found! Using default font.")
    sin_font = None  # Falls back to default font

def describe_corpus(file_path):
    """
    Compute and print basic corpus statistics.
    """
    num_lines = 0
    num_chars = 0
    word_counts = []
    unique_words = set()

    with open(file_path, "r", encoding="utf-8") as f:
        for line in f:
            num_lines += 1
            num_chars += len(line)
            words = line.strip().split()
            word_counts.extend(words)
            unique_words.update(words)

    file_size_mb = os.path.getsize(file_path) / (1024 * 1024)

    print("\n[CORPUS STATISTICS]")
    print(f"Total lines: {num_lines}")
    print(f"Total characters: {num_chars}")
    print(f"Total words: {len(word_counts)}")
    print(f"Unique words: {len(unique_words)}")
    print(f"File size: {file_size_mb:.2f} MB\n")

    return word_counts

def print_top_words(word_counts, top_n=10):
    """
    Print the top N most common words in the corpus.
    """
    word_freq = Counter(word_counts)
    common_words = word_freq.most_common(top_n)

    print("\n--- TOP 10 WORDS ---")
    for word, freq in common_words:
        print(f"{word}: {freq}")

def plot_word_frequency(word_counts, top_n=30):
    """
    Plot the most common words in the corpus with Sinhala font.
    """
    word_freq = Counter(word_counts)
    common_words = word_freq.most_common(top_n)

    words, freqs = zip(*common_words)

    plt.figure(figsize=(12, 6))
    sns.barplot(x=list(words), y=list(freqs))
    plt.xticks(rotation=45, fontproperties=sin_font if sin_font else None)  # Apply Sinhala font if available
    plt.xlabel("Words", fontproperties=sin_font if sin_font else None)
    plt.ylabel("Frequency", fontproperties=sin_font if sin_font else None)
    plt.title(f"Top {top_n} Most Common Words in Sinhala Corpus", fontproperties=sin_font if sin_font else None)
    plt.show()

def analyze_sinhala_characters(word_counts, top_n=10):
    """
    Analyze the presence of Sinhala-specific characters in the corpus and print top characters.
    """
    sinhala_chars = Counter()

    for word in word_counts:
        for char in word:
            sinhala_chars[char] += 1

    # Sort characters by frequency
    sorted_chars = sorted(sinhala_chars.items(), key=lambda x: x[1], reverse=True)

    # Print top N Sinhala characters
    print("\n--- TOP 10 CHARACTERS ---")
    for char, count in sorted_chars[:top_n]:
        print(f"{char}: {count}")

    # Plot character frequency
    plt.figure(figsize=(12, 5))
    chars, counts = zip(*sorted_chars[:30])
    sns.barplot(x=list(chars), y=list(counts))
    plt.xticks(fontproperties=sin_font if sin_font else None)  # Apply Sinhala font if available
    plt.xlabel("Character", fontproperties=sin_font if sin_font else None)
    plt.ylabel("Frequency", fontproperties=sin_font if sin_font else None)
    plt.title("Most Common Sinhala Characters in Corpus", fontproperties=sin_font if sin_font else None)
    plt.show()

if __name__ == "__main__":
    # Run EDA
    if os.path.exists(CORPUS_FILE):
        word_counts = describe_corpus(CORPUS_FILE)
        print_top_words(word_counts)  # Print top words
        plot_word_frequency(word_counts)
        analyze_sinhala_characters(word_counts)  # Print top characters
    else:
        print(f"[ERROR] Corpus file '{CORPUS_FILE}' not found.")
```
